```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Part 2 - Regression/Section 6 - Polynomial Regression/Python/Position_Salaries.csv')
```

```python
x = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.preprocessing import StandardScaler

scx = StandardScaler()
scy = StandardScaler()
y=y.reshape(len(y),1)
x = scx.fit_transform(x)
y = scy.fit_transform(y)

```

```python
from sklearn.svm import SVR

sv = SVR(kernel = 'rbf')
sv.fit(x,y)
```

```python
z = scy.inverse_transform(sv.predict(scx.transform([[6.5]])).reshape(-1,1))
print(z[0][0])
```

```python
import matplotlib.pyplot as plt

plt.scatter(scx.inverse_transform(x),scy.inverse_transform(y))
plt.plot(scx.inverse_transform(x),scy.inverse_transform(sv.predict(x).reshape(-1,1)))
```

```python
x_grid = np.arange(min(scx.inverse_transform(x)),max(scx.inverse_transform(x)),0.1)
x_grid = x_grid.reshape(len(x_grid),1)
plt.scatter(scx.inverse_transform(x),scy.inverse_transform(y))
plt.plot(x_grid,scy.inverse_transform(sv.predict(scx.transform(x_grid)).reshape(-1,1)))
```
